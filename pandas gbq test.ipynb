```python
from google.cloud import bigquery
import os
import pandas
from pandas.io import gbq

projectID = 'andrewtest-208508'

#queries and dataframes
query_age = 'SELECT * FROM z_andrew_sandbox.profiles_with_age'
age_df = pandas.read_gbq(query=query_age, project_id= projectID)


query_edu = 'SELECT * FROM z_andrew_sandbox.education_names2'
edu_df= pandas.read_gbq(query=query_edu, project_id=projectID)

query_work = 'Select asid_h, employer_name, location_name from z_andrew_sandbox.work_2'
work_df = pandas.read_gbq(query=query_work, project_id=projectID)

query_likes = 'Select * from z_andrew_sandbox.likes_names'
likes_df = pandas.read_gbq(query=query_likes, project_id = projectID)



#merging>
combined_df=pandas.merge(age_df,edu_df,on='asid_h')
combined_df=combined_df.drop_duplicates()

combined_df=pandas.merge(combined_df, work_df, on='asid_h')
combined_df=combined_df.drop_duplicates()

combined_df=pandas.merge(combined_df, likes_df, on='asid_h')
combined_df=combined_df.drop_duplicates()






```
